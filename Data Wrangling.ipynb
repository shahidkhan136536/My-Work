# Data Normalization
 to uniform the data means range is made easy where our data is more clear to make a graph

```python
kashti.head()
```

```python
ks4 = kashti[["age in days", 'fare']]
ks4.head()
```

```python
#Simple scaling feature
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4.head()
```

```python
# Min - Max method
ks4['fare'] = (ks4['fare']-ks4['fare'].min()) / (ks4['fare'].max()-ks4['fare'].min())
ks4.head()
```

```python
# Z-score (standard score) # range from 0 to 3
ks4['fare'] = (ks4['fare']-ks4['fare'].mean()) / ks4['fare'].std()
ks4.head()
```
